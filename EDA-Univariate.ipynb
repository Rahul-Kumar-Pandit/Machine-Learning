```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
df.sample(5)
```

# **1. Categorical Cols**

## **a.countplot (bar chart)** 

```python
sns.countplot(x=df['SibSp'])
```

## **b. PieChart**

```python
df['Survived'].value_counts()
```

```python
df['Survived'].value_counts().plot(kind='pie',autopct='%.2f')
```

# **2. Numerical Data Cols**

## **a. Histogram**

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'])
```

## **b.Distplot**

```python
sns.histplot(df["Age"], kde=True, stat="probability")
```

## **c. Boxplot**

```python
sns.boxplot(x=df['Age'])
```
